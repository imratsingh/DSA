# Recursion function

```python
n = 5
i = 1
while i<=n:
    print(i,end=" ")
    i+=1
```

```python
def printnumber(i,n):
    #base case
    if i >n:
        return
    #recursive case
    print(i,end=" ")
    printnumber(i+1,n)
printnumber(1,5)
```

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(4))
```
