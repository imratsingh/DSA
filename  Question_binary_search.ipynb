# Q1. peak index in a mountain array

```python
def peakindex(nums):
    n = len(nums)
    l = 0
    r = n-2
    ans = n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid] < nums[mid+1]:
            l = mid+1
        else:
            ans = mid
            r = mid-1
    return ans
print(peakindex([1,2,3,5,3,4]))
```

# Q2. Square number

```python
def number(x):
    if x == 0:
        return 0
    l = 1
    r = x
    ans = 1
    while l<=r:
        mid = (l+r)//2
        midsquare = mid*mid
        if midsquare>x:
            r = mid-1
        else:
            ans = mid
            l = mid+1
    return ans
print(number(20))
```

# Q.3 Koko eating bananas

```python
def getHours(piles,mid):
    ans = 0
    for pile in piles:
        ans += (pile+mid-1)/mid
    return ans
    
def mineatingspeed(piles,h):
    n = len(piles)
    l = 1
    r = max(piles)
    k = r
    while l<=r:
        mid = (l+r)//2 
        if getHours(piles,mid)>h:
            l = mid+1
        else:
            k = mid
            r = mid-1
    return k

print(mineatingspeed([3,6,7,11],8))
```
