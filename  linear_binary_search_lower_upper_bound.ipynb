# Linear Search

```python
Linear Search is a searching algorithm in which elements are checked
one by one until the target element is found.
```

```python
def linearsearch(nums,target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1


def search(nums):
    
    target = int(input("Enter your target number :"))
    p = linearsearch(nums,target)
    return p
    
print(search([23,34,45,56,67,789]))
```

# Binary Search

```python
def binarysearch(nums,target):
    n = len(nums)
    l = 0
    r = n-1

    while l<=r:
        mid = (l+r)//2
        if target == nums[mid]:
            return mid
        elif target > nums[mid]:
            # right 
            l = mid+1
        else:
            # left
            r = mid-1
    return -1

print(binarysearch([1,2,3,4,5,6,7,8,9,],8))
```

# lower bound

```python
def lowerbound(nums,target):
    n = len(nums)
    l = 0
    r = n-1
    while l<=r:
        mid = (l+r)//2
        if nums[mid] == target:
            ans = mid
            r = mid-1
        else:
            #right
            l = mid+1
    return ans
print(lowerbound([1,2,3,4,5,6,7,8],6))
```

# upperbound

```python
def upperbound(nums,target):
    n = len(nums)
    l = 0
    r = n-1
    ans = n
    while l<=r:
        mid = (l+r)//2
        if nums[mid] > target:
            ans = mid
            r = mid-1
        else:
            l = mid+1
    return ans
print(upperbound([1,2,3,4,5,6,7,8],6))
```
