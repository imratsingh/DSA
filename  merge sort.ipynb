# merge sort

Merge Sort is a Divide and Conquer sorting algorithm.
    
Merge Sort is a sorting algorithm that divides an array into smaller parts,
sorts them recursively, and merges them into a sorted array.

```python
def merge(nums,l,mid,r):
    a = []
    b = []
    
    for i in range(l,mid+1):
        a.append(nums[i])
        
    for i in range(mid+1,r+1):
        b.append(nums[i])
        
    i,j,k = 0,0,1
    
    while k<=r:
        
        if j==len(b):
            nums[k] = a[i]
            i +=1
            k +=1
            
        elif i==len(a):
            nums[k] = b[i]
            j +=1
            k +=1
            
        elif a[i] < b[j]:
            nums[k] = a[i]
            i +=1
            k +=1
            
        else:
            nums[k] = b[j]
            j +=1
            k +=1
            
def mergesort(nums,l,r):
    if l>=r:
        return
        
    mid = (l+r)//2
    
    mergesort(nums,l,mid+1)
    mergesort(nums,mid+1,r)
    
    merge(nums,l,mid,r)
def sortarray(nums):
    mergesort(nums,0,len(nums)-1)
    return nums


print(sortarray([23,45,56,57,67,78,89,1,2,3,56,0]))
```
